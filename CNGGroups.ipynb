```python
import sqlite3
import pandas as pd
import json
```

```python
#These Json Files contain the CNG Groupings 
dictnoun = json.load(open("noungroups.json"))
dictverb = json.load(open("verbgroups.json"))
```

```python
conn = sqlite3.connect("Extracted.db")
```

```python
df = pd.DataFrame(columns = ["CNG", "CNG_Group"])
#defining a dataframe in which we have the list of all CNG_Groups to which a CNG Belongs
#We define it in such a way because we want to use it in an inner join to add CNG_Group nodes in the graph.
```

```python
df
```

```python
z =0
#z = index of the dataframe

for key in dictnoun.keys():
    #key is a CNG_Group
    for value in dictnoun[key]:
        #value is CNG
        df.loc[z] = [value, key]
        #storing CNG and its GROUP
        z += 1
        #incrementing index
        
```

```python
df.head()
```

```python
#same as above but for verbs groups
for key in dictverb.keys():
    for value in dictverb[key]:
        df.loc[z] = [(-1)*value, key]
        #-1 is multipled as CNG of Verbs are negative.
        z += 1
        
```

```python
df
```

```python
df.to_sql("CNG&CNGG", conn)
#dumping data in a table
```

```python
conn.close()
```

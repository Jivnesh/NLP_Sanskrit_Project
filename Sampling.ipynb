```python
import sqlite3 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("Test.db")
```

```python
#Wordgraph is a subset of Graph, it only contains word -> word edges.
#While considering our sample , we are selecting only word nodes as external nodes in our metapaht
#We only use words as they contaiin inforamtion of both Lemma and CNG
df =  pd.read_sql("SELECT * FROM WordGraph", conn)
print(df.head())
```

```python
df.describe()
```

```python
ft = df.groupby(['Weight']).agg(['count'])
```

```python
print(ft)
```

```python
dfs = df.sort_values( by = ['Weight'], ascending = False) 
```

```python
dfs = dfs.reset_index(drop = True)
```

```python
dfs = dfs.reset_index(drop = False)
```

```python
dfs
```

```python
sns.lmplot(x = 'index', y = 'Weight', data = dfs)
plt.show()
```

```python
#max = 368
#min = 1
#number = 132609
#from the plot we can see that we should consider weights below 100
dfs = dfs.loc[dfs['Weight'] < 100 ]
```

```python
import pickle         #used to read pickle instances of an object
import sqlite3        #for sqlite databases/local databases
import pandas as pd   #for general database operations
import threading      #for parallel processing
```

```python
#pickles are an instance of class DCS

class DCS:
    def __init__(self, sent_id, sentence):
        self.sent_id = sent_id
        self.sentence = sentence
        self.dcs_chunks = []
        self.lemmas = []
        self.cng = []              
```

```python
#function to extract data and dump in a .db file. 
#Index1 is the index of a particular lemma in a sentence
#Index2 is the index of a particulat chunk in that lemma
#next time make tables in the same db file not diffrent files!

def extraction(start, end,) :
    if start == 0 : 
        start = 1 # for proper indexing
    if start == 364900 :
        start = start+1
    conn = sqlite3.connect(str(start)+'.db')  #creating a connection to local database
    df = pd.DataFrame(columns = ['Filename', 'Lemma', 'CNG', 'Index1', 'Index2'])
    print(df.head())
    z=0
    
    for i in range(start,end):
        if i > 442383:
            break
        
        try:
            output_load = pickle.load(open("C:\\Users\\Mooga\\Documents\\NLP Project\\DCS_pick" +"\\" + str(i) + '.p', "rb"), encoding='utf-8')
           
            for j in range(len(output_load.lemmas)):
                
                for k in range(len(output_load.lemmas[j])):
                    df.loc[z]= [i, output_load.lemmas[j][k], output_load.cng[j][k], j, k]
                    z = z+1  
        except:
            pass
        if i%100 == 0 :
            print(i,j,k)
        
    df.to_sql(str(start)+'.db', conn)
    conn.close()
    print(start, ' to ', end)

    
  
```

```python
#threading the process to increase speed/ parallel processing
#no of files = 442383
w = 0
y = 2225
itr = 20
for x in range(itr):
    
    t1 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t2 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t3 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t4 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t5 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t6 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t7 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t8 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t9 = threading.Thread(target=extraction, args=(w,w+y))
    w += y
    t10 = threading.Thread(target=extraction, args=(w,w+y))
    w += y

                        
    t1.start()
    t2.start()
    t3.start()
    t4.start()
    t5.start()
    t6.start()
    t7.start()
    t8.start()
    t9.start()
    t10.start()

    t1.join()
    t2.join()
    t3.join()
    t4.join()
    t5.join()
    t6.join()
    t7.join()
    t8.join()
    t9.join()
    t10.join()
```

```python
import pandas as pd 
import sqlite3
import json
from collections import defaultdict
import pickle
```

```python
#These Json Files contain the CNG Groupings 
dictnoun = json.load(open("noungroups.json"))
dictverb = json.load(open("verbgroups.json"))
```

```python
#again, run it on Extracted.db in the server
conn = sqlite3.connect("Test.db")
#conn = sqlite3.connect("Extracted.db")
```

```python
df = pd.read_sql("SELECT * FROM Extracted", conn)
```

```python
df.head()
```

```python
d = defaultdict(set)
#we use a dictonary of sets as we want count of distinct filenames, so taking union is easy
dLemma = defaultdict(set)
dWord = defaultdict(set)
```

```python
for r, i in df.iterrows():
    #iterating through each Lemma/CNG/word in Each scentence
    d[i['CNG']].add(i['Filename'])
    #since they are sets we will get only distinct values
    dLemma[i['Lemma']].add(i['Filename'])
    dWord[i['Word']].add(i['Filename'])

print(d)
```

```python
CNGG_count = {}
for key in dictnoun.keys() :
    #iterating thorugh all CNG_groups of type nouns
    uni = set()
    #now we take the union of all the files which CNG's of a CNG Group occur in
    for i in dictnoun[key]:
        #iterating though CNG's in CNG_Group
        uni = uni.union(d[i])
    CNGG_count[key] = len(uni)
    
    
for key in dictverb.keys() :
    uni = set()
    for i in dictverb[key]:
        uni = uni.union(d[-i])
    CNGG_count[key] = len(uni)
 
```
